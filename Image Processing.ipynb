```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
img_path = r'C:\Users\fabio\Documents\foto2.jpg'
```

```python
img = cv2.imread(img_path)
```

```python
img.shape
```

```python
sns.set(rc={'figure.figsize': (10, 6)})
```

```python
def show_original(img):
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original')
    plt.axis('off')
```

```python
show_original(img)


plt.subplot(1, 2, 2)
threshold = 160
img[img>threshold]=255
img[img<threshold]=0.0
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.title("Thresholded")
plt.axis('off')

plt.show()
```

```python
img = cv2.imread(r'C:\Users\fabio\Pictures\Saved Pictures\ships.jpg', 0)
```

```python

show_original(img)

kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

sns.set(rc={'figure.figsize' : (15, 10)})
plt.subplot(1, 2, 2)
conv = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(conv, cv2.COLOR_BGR2RGB))
plt.title('Edge Detection with Convolution')
plt.axis('off')
plt.show()
```
