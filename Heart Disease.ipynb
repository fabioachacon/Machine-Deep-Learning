```python
#Separates the examples from the ground truth labels.
data, data_labels = df.iloc[1:, :-1].values, df.iloc[1:, -1:].values
```

```python
#Split the data into a training set and a test set (30% of the data is destined to test the performance of the model).
x_train, x_test, y_train, y_test = train_test_split(data, data_labels, test_size=0.30, stratify=data_labels)
```

The next cells compare the performance of three classifiers.

```python
#Standardizes the data for optimal gradient-based optimization.
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
#Plots a validation curve.
def val_curve(est, x, y, par_name):
    param_range=[0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
    train_scores, test_scores = validation_curve(estimator=est,
                                                 X=x,
                                                 y=y.ravel(),
                                                 param_name=par_name,
                                                 param_range=param_range,
                                                 cv=10)
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    plt.plot(param_range, train_mean, color='blue', marker='o', markersize=5, label='training accuracy')
    plt.fill_between(param_range, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
    plt.plot(param_range, test_mean, color='red', marker='s', linestyle='--', markersize=5, label='test accuracy')
    plt.fill_between(param_range, test_mean + test_std, test_mean - test_std, alpha=0.15)
    
    plt.grid()
    plt.xscale('log')
    plt.xlabel('Parameter C')
    plt.ylabel('Accuracy')
    plt.ylim([0.8, 1.03])
    plt.show()
```

```python
#Fits the model with Logistic Regression.
lr = LogisticRegression(C=100, random_state=0)
lr.fit(x_train_std, y_train.ravel())
y_pred = lr.predict(x_test_std)
print('Accuracy: %.2f' % accuracy_score(y_test.ravel(), y_pred))
```

```python
joblib.dump(lr, 'logregressionmodel.pkl') 
```

```python
pipe_lr = make_pipeline(StandardScaler(), lr)
```

```python
#The information provided by this curve is used as feedback for hyperparameter tuning.
val_curve(est=pipe_lr, x=x_train, y=y_train, par_name='logisticregression__C')
```

```python
#Fiting the model using Naive Bayes.
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train_std, y_train.ravel())
y_pred = model.predict(x_test_std)
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

```python
#Fits a SVM model with a nonlinear Kernel.
from sklearn.svm import SVC
svm = SVC(kernel='rbf', random_state=0, C=3.0)
svm.fit(x_train_std, y_train.ravel())
y_pred = svm.predict(x_test_std)
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

```python
#Reduces the Dimension via PCA.
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
x_train_pca = pca.fit_transform(x_train_std)
x_test_pca = pca.transform(x_test_std)
lr.fit(x_train_pca, y_train.ravel())
y_pred_pca = lr.predict(x_test_pca)
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred_pca))
```
