```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = r'C:\Users\fabio\Documents\foto2.jpg'
```

```python
img = cv2.imread(img_path)
```

```python
img.shape
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')
plt.show()

threshold = 160
img[img>threshold]=255
img[img<threshold]=0.0
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Thresholded")
plt.show()
```

```python
img = cv2.imread(r'C:\Users\fabio\Pictures\Saved Pictures\ships.jpg', 0)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')
plt.show()

kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

conv = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(conv, cv2.COLOR_BGR2RGB))
plt.title('Edge Detection with Convolution')
plt.axis('off')
plt.show()
```

```python
img_path = r'C:\Users\fabio\Pictures\beethoven1.jpg'
image = cv2.imread(img_path)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')
plt.show()


kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1,-0]])

conv = cv2.filter2D(image, -1, kernel)
plt.imshow(cv2.cvtColor(conv, cv2.COLOR_BGR2RGB))
plt.title('Sharpen')
plt.axis('off')
plt.show()
```
