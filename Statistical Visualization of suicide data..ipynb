```python
df = pd.read_csv(r'C:\Users\fabio\Documents\DataSets\csv_data\master.csv')
```

```python
categories_a = df.groupby(['sex'])['suicides_no'].sum().index.tolist()
categories_b = df.groupby(['age'])['suicides_no'].sum().index.tolist()
categories_c = df.groupby(['country'])['suicides_no'].sum().nlargest(6).index.tolist()
categories_d = df.groupby(['generation'])['suicides_no'].sum().index.tolist()
```

```python
from matplotlib.colors import LogNorm
from numpy import median

_, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10), sharex=True)
columns = ['sex', 'age', 'country', 'generation']
categories = [categories_a, categories_b, categories_c, categories_d]
index = [axes[0, 0], axes[0, 1], axes[1, 0], axes[1, 1]]
for col, cat, pos in zip(columns, categories, index):
    data = df.groupby(['year', col])['suicides_no'].sum().reset_index()
    data[col] = data[col].astype(pd.api.types.CategoricalDtype(categories=cat))
    cond = col in ['generation', 'country']
    ax = sns.lineplot(x='year', y='suicides_no', hue=col, style=col, hue_norm=LogNorm(), estimator=median, data=data_g, sort=False, dashes=cond, ax=pos)
    ax.set_ylabel("Number of Suicides")

plt.tight_layout()
plt.show()    
```

```python
dt = df.groupby(['country', 'year']).sum().suicides_no
```

```python
def trace_plot(data, categories, set_mode):
    traces = []
    for cat in categories:
        tr = go.Scatter(x=dt[cat].reset_index().year.values,
                        y=dt[cat].reset_index().suicides_no.values,
                        mode=set_mode,
                        name=cat)
        traces.append(tr)
    return py.iplot(traces, filename='line-mode')
```

```python
trace_plot(data=df, categories=categories_c, set_mode='lines+markers')
```
